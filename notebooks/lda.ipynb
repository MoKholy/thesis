```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

# import linear discriminant analysis
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.preprocessing import StandardScaler

# append scripts to system path
import sys
sys.path.append("../scripts/")
from utils import split_data, plot_2d_data
```

```python
# load parqet file
df = pd.read_parquet("../data/correctly_labeled.parquet")
```

```python
data_x, data_y, data_y_string = split_data(df)
```

```python
# perform lda on the data

lda = LinearDiscriminantAnalysis(n_components=2)
x_lda = lda.fit_transform(data_x, data_y_string)

```

```python
# plot data
plot_2d_data(x_lda, data_y_string, "LDA 2D plot")
```
